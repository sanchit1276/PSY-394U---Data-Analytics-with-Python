```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Problem 1
```

```python
#import data
sw = pd.read_csv('Southwest_Mar2013.csv')
```

```python
departing = sw.groupby('ORIGIN').sum()
departing.columns = ['Departing']
```

```python
arriving = sw.groupby('DEST').sum()
arriving.columns = ['Arriving']
```

```python
NumPassengersByAirport = pd.concat([departing, arriving], axis=1, sort=True)
NumPassengersByAirport = NumPassengersByAirport.fillna(0)
```

```python
NumPassengersByAirport.to_csv('NumPassengersByAirport.csv',index=True)
```

```python
# plotting
NumPassengersByAirport.plot.scatter(x='Departing', y='Arriving', title='Number of Departing vs Arriving Passengers for all airports')
```

```python
# top 5 airports for departures
print(departing.sort_values(by=['Departing'],ascending=False).head(5))
```

```python
#import data
air = pd.read_csv('AllAirlines_Mar2013.csv')
```

```python
tot_departures = air.groupby('ORIGIN')['PASSENGERS'].sum()
tot_departures = tot_departures.to_frame()
```

```python
air = air[air.UNIQUE_CARRIER_NAME == 'Southwest Airlines Co.']
departures_sw = air.groupby('ORIGIN')['PASSENGERS'].sum()
departures_sw = departures_sw.to_frame()
```

```python
SW_share = departures_sw / tot_departures
SW_share.columns = ['Southwest']
SW_share = SW_share.fillna(0)
```

```python
SW_share.to_csv('SouthwestShare.csv',index=True)
```
